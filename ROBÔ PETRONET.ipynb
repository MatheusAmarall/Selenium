### Robô da Petronet

### Desafio

Tinhamos uma prestação de contas a ser feita no site da Petrobras, que era realizado manualmente, pelo setor de faturamento, segundo a Lyssa, era algo crítico, que tomava muito tempo, e era realizado pelos 5 colaboradores que faturavam as diferentes regiões.<br>
Como solução, construimos um robô por selenium, e que apenas com uma planilha, que já era feita pelo setor, e que teve apenas algumas informações adicionadas, ele realiza o cadastro com muita agilidade (cerca de 1 minuto por cadastro), e também trouxe a oportunidade dos colaboradores realizarem outras atividades nesse tempo que eles teriam que gastar fazendo esses cadastros. 

### Entrando no site da Petrobras

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.keys import Keys
import time
import pandas as pd
import pyautogui
import os
import autoit
from pathlib import Path

lista_notas = os.listdir('Notas')

#entrar no site da prefeitura
driver = webdriver.Chrome()
driver.get('https://www.petronect.com.br/irj/portal/anonymous')

time.sleep(2)
#fazer login no site
driver.find_element(By.CLASS_NAME, 'login').click()
time.sleep(3)

#input login
driver.find_element(By.XPATH, '//*[@id="inputUser"]').send_keys('MM.EQUISUL')
#input senha
driver.find_element(By.XPATH, '//*[@id="inputSenha"]').send_keys('Eqs80464')
#entrar
driver.find_element(By.XPATH, '//*[@id="Continuar"]').click()

time.sleep(3)

#clicar em pagamentos
driver.find_element(By.XPATH, '//*[@id="tabIndex4"]').send_keys(Keys.RETURN)
time.sleep(2)
```

```python
df = pd.read_excel(r'PLANILHA PETROBRAS 21.01.2022.xlsx')
df = df.set_index('INDICE')


nome_arquivos = os.listdir('Notas')



for nome in nome_arquivos:
    #voltar para a pagina normal
    driver.switch_to.default_content()
    time.sleep(2)
    ####clicar em carregar documentos fiscais
    driver.find_element(By.XPATH, '//*[@id="subTabIndex2"]/div[1]').click()
    time.sleep(5)

    #entrar no primeiro iframe
    iframe = driver.find_element(By.XPATH, '//*[@id="contentAreaFrame"]')
    driver.switch_to.frame(iframe)
    #entrar no segundo iframe
    iframe = driver.find_element(By.XPATH, '//*[@id="isolatedWorkArea"]')
    driver.switch_to.frame(iframe)


    #documento fiscal eletronico (SEMPRE)
    while len(driver.find_elements(By.XPATH, '//*[@id="WD2D"]')) == 0:
        time.sleep(2)

    driver.find_element(By.XPATH, '//*[@id="WD2D"]').click()
    time.sleep(2)

    #escolher o modelo 56
    driver.find_element(By.XPATH, '//*[@id="WD61"]').click()
    time.sleep(2)
    
    ####INSERINDO OS ANEXOS
    
    #carregar documento
    lista_arquivos = r'C:\Users\USER\Desktop\Jupyter\Robo Petrobras\Notas\{}'.format(nome)
    time.sleep(2)

    #clicar no escolher arquivo (aberto para melhorias)
    pyautogui.click(x=1104, y=729)
    time.sleep(2)

    #Carregar documento
    autoit.win_active("Abrir") 
    autoit.control_send("Abrir","Edit1",r"{}".format(lista_arquivos))
    autoit.control_send("Abrir","Edit1","{ENTER}")  
    time.sleep(2)

    #carregar anexo 
    driver.find_element(By.XPATH, '//*[@id="WD99"]').send_keys(Keys.RETURN)
    time.sleep(3)
    
    #voltar para a pagina normal
    driver.switch_to.default_content()
    time.sleep(3)

    #entrar no iframe para escolher o arquivo
    iframe = driver.find_element(By.XPATH, '//*[@id="URLSPW-0"]')
    driver.switch_to.frame(iframe)
    time.sleep(2)

    #clicar no escolher arquivo (aberto para melhorias)
    pyautogui.click(x=1117, y=426)
    time.sleep(2)

    #escolher arquivo
    autoit.win_active("Abrir") 
    autoit.control_send("Abrir","Edit1",r"{}".format(lista_arquivos))
    autoit.control_send("Abrir","Edit1","{ENTER}")  
    time.sleep(2)

    #clicar em ok
    driver.find_element(By.CLASS_NAME, 'lsButton--base').click()
    time.sleep(2)
    
    ####NOTA FISCAL REFERENCIADA
    
    #voltar para a pagina normal
    driver.switch_to.default_content()
    time.sleep(3)

    #entrar no primeiro iframe
    iframe = driver.find_element(By.XPATH, '//*[@id="contentAreaFrame"]')
    driver.switch_to.frame(iframe)
    #entrar no segundo iframe
    iframe = driver.find_element(By.XPATH, '//*[@id="isolatedWorkArea"]')
    driver.switch_to.frame(iframe)

    #nota fiscal referenciada
    driver.find_element(By.XPATH, '//*[@id="WDE8"]').click()

    #voltar para a pagina normal
    driver.switch_to.default_content()
    time.sleep(3)

    #entrar no iframe para escolher o arquivo
    iframe = driver.find_element(By.ID, 'URLSPW-0')
    driver.switch_to.frame(iframe)
    time.sleep(3)

    #digitar o número da nota
    nota = int(df['NOTA'][int(nome[:-4])])
    driver.find_element(By.CLASS_NAME, 'lsField__input').send_keys(nota)
    time.sleep(3)

    #enter
    driver.find_elements(By.CLASS_NAME, 'lsButton')[0].click()
    time.sleep(2)
    
    #####INFORMAÇÕES ADICIONAIS
    
    #voltar para a pagina normal
    driver.switch_to.default_content()
    time.sleep(3)

    #entrar no primeiro iframe
    iframe = driver.find_element(By.XPATH, '//*[@id="contentAreaFrame"]')
    driver.switch_to.frame(iframe)
    #entrar no segundo iframe
    iframe = driver.find_element(By.XPATH, '//*[@id="isolatedWorkArea"]')
    driver.switch_to.frame(iframe)
    time.sleep(2)

    #informações adicionais
    driver.find_element(By.XPATH, '//*[@id="WD010B"]').click()
    time.sleep(3)

    #voltar para a pagina normal
    driver.switch_to.default_content()
    time.sleep(3)

    #entrar no iframe para escolher o arquivo
    iframe = driver.find_element(By.ID, 'URLSPW-0')
    driver.switch_to.frame(iframe)
    time.sleep(6)

    #numero da nota fiscal
    driver.find_elements(By.CLASS_NAME, 'lsField__input')[0].send_keys(nota)
    time.sleep(1.5)
    
    #documento de obrigação
    driver.find_element(By.CLASS_NAME, 'lsSelector--generic').click()
    time.sleep(1.5)

    #clicar em inserir linhas
    driver.find_element(By.CSS_SELECTOR, 'div.lsButton--base').send_keys(Keys.RETURN)
    time.sleep(2)

    #número da FRS
    FRS = int(df['FR'][int(nome[:-4])])
    driver.find_elements(By.CLASS_NAME, 'lsField__input')[1].send_keys(FRS)
    time.sleep(1.5)

    #clicar em inserir linhas
    driver.find_elements(By.CSS_SELECTOR, 'div.lsButton--base')[1].send_keys(Keys.RETURN)
    time.sleep(2)

    #número da NL
    NL = int(df['NL'][int(nome[:-4])])
    driver.find_elements(By.CLASS_NAME, 'lsField__input')[2].send_keys(NL)
    time.sleep(1.5)

    #data
    DATA = str(df['DATA EMISSAO'][int(nome[:-4])])
    driver.find_elements(By.CLASS_NAME, 'lsField__input')[3].send_keys(DATA)
    time.sleep(1.5)

    #uf
    UF_TOMADOR = str(df['UF TOMADOR'][int(nome[:-4])])
    driver.find_elements(By.CLASS_NAME, 'lsField__input')[4].send_keys(UF_TOMADOR)
    time.sleep(1.5)

    #municipio brota
    MUNICIPIO_TOMADOR = str(df['MUNICIPIO TOMADOR'][int(nome[:-4])])
    driver.find_elements(By.CLASS_NAME, 'lsField__input')[5].send_keys(MUNICIPIO_TOMADOR)
    time.sleep(1.5)

    #item lei comp. 116
    PRODUTO = str(df['PRODUTO'][int(nome[:-4])])
    if PRODUTO == '7.1':    
        driver.find_elements(By.CLASS_NAME, 'lsField__input')[6].send_keys('07.10')
    else:
        driver.find_elements(By.CLASS_NAME, 'lsField__input')[6].send_keys('14.01')
    time.sleep(1.5)

    #serviços realizos em:
    driver.find_elements(By.CLASS_NAME, 'lsField__input')[7].send_keys('Terras')
    time.sleep(1.5)

    #dependencias
    driver.find_elements(By.CLASS_NAME, 'lsField__input')[8].send_keys('Dependências da Petrobras')
    time.sleep(1.5)

    #uf
    UF_PRESTADOR = str(df['UF PRESTADOR'][int(nome[:-4])])
    driver.find_elements(By.CLASS_NAME, 'lsField__input')[9].send_keys(UF_PRESTADOR)
    time.sleep(2)

    #municipio de execução
    MUNICIPIO_PRESTADOR = str(df['MUNICIPIO PRESTADOR'][int(nome[:-4])])
    driver.find_elements(By.CLASS_NAME, 'lsField__input')[10].send_keys(MUNICIPIO_PRESTADOR)
    time.sleep(1.5)
    
    
    ##IR
    IR = str(df['IR'][int(nome[:-4])])
    driver.find_elements(By.CLASS_NAME, 'lsField__input')[11].send_keys(IR.replace(".", ","))
    time.sleep(1)

    ##INSS
    INSS = str(df['INSS'][int(nome[:-4])])
    driver.find_elements(By.CLASS_NAME, 'lsField__input')[12].send_keys(INSS.replace(".", ","))
    time.sleep(1)

    ##PIS
    PIS = str(df['PIS'][int(nome[:-4])])
    driver.find_elements(By.CLASS_NAME, 'lsField__input')[13].send_keys(PIS.replace(".", ","))
    time.sleep(1)

    ##COFINS
    COFINS = str(df['COFINS'][int(nome[:-4])])
    driver.find_elements(By.CLASS_NAME, 'lsField__input')[15].send_keys(COFINS.replace(".", ","))
    time.sleep(1)

    ##ISS
    ISS = str(df['ISS'][int(nome[:-4])])
    driver.find_elements(By.CLASS_NAME, 'lsField__input')[16].send_keys(ISS.replace(".", ","))
    time.sleep(1)

    ##CSLL
    CSLL = str(df['CSLL'][int(nome[:-4])])
    driver.find_elements(By.CLASS_NAME, 'lsField__input')[17].send_keys(CSLL.replace(".", ","))
    time.sleep(1)
    
    ###Valor total
    #limpar
    driver.find_elements(By.CLASS_NAME, 'lsField__input')[23].clear()
    time.sleep(1.5)
    #inputar
    VALOR = str(df['VALOR'][int(nome[:-4])])
    driver.find_elements(By.CLASS_NAME, 'lsField__input')[23].send_keys(VALOR.replace(".", ","))
    time.sleep(1.5)

    ###Moeda
    driver.find_elements(By.CLASS_NAME, 'lsField__input')[24].send_keys('Real - Brasil')
    time.sleep(1.5)

    ###Ok
    driver.find_elements(By.CLASS_NAME, 'lsButton--base')[2].click()
    time.sleep(2)
    
    
    ####ENVIAR
    
    #voltar para a pagina normal
    driver.switch_to.default_content()
    time.sleep(3)

    #entrar no primeiro iframe
    iframe = driver.find_element(By.XPATH, '//*[@id="contentAreaFrame"]')
    driver.switch_to.frame(iframe)
    #entrar no segundo iframe
    iframe = driver.find_element(By.XPATH, '//*[@id="isolatedWorkArea"]')
    driver.switch_to.frame(iframe)
    time.sleep(2)

    #clicar em enviar
    driver.find_element(By.XPATH, '//*[@id="WD0114"]').click()
    time.sleep(3)
    
    ###após enviar clicar em ok, precisa entrar nos 2 iframes
    #voltar para a pagina normal
    driver.switch_to.default_content()
    time.sleep(2)
    #entrar no iframe para escolher o arquivo
    iframe = driver.find_element(By.ID, 'URLSPW-0')
    driver.switch_to.frame(iframe)
    time.sleep(2)
    #clicar no ok
    driver.find_element(By.CLASS_NAME, 'lsButton').click()
    time.sleep(2)

pyautogui.alert('O código terminou de rodar')

    
```
